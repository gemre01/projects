```python
import requests
import pandas as pd
import csv
```

```python
files = ["players.csv", "teams.csv"]

for file in files:
    CSV_URL = "https://raw.githubusercontent.com/dataquestio/project-walkthroughs/master/mvp/{}".format(file)

    data  = []
    with requests.get(CSV_URL, stream=True) as r:
        lines = (line.decode('utf-8') for line in r.iter_lines())
        for row in csv.reader(lines):
            data.append(row)
    
    df = pd.DataFrame(data)
    df.columns = df.iloc[0, :]
    df = df.drop(df.index[0])

    df.to_csv("{}".format(file))
```
